# Installing Magenta using Python 3.7


**Step 1:** 



> WINDOWS-- Install Ubuntu and change Windows subsytems ([link here](https://www.onmsft.com/how-to/install-ubuntu-on-windows-10-or-windows-11))

> MAC-- Open terminal and switch shell from Zsh to Bash, see below

```
# Run in shell
chsh -s /bin/bash

```

**Step 2:** Install Anaconda if you don't already have it installed. The installation process for windows is [here](https://docs.anaconda.com/anaconda/install/windows/). The installation process for MacOS is [here](https://docs.anaconda.com/anaconda/install/mac-os/).


> **Note:** When installing on Windows, check the add to path option when presented the screen below.

win-install-options.png

**Step 3:** Attempt auto install in Linux Environment.

  **Auto-install:** Open your powershell.

```
#Ensure Ubuntu or other Linux OS is running on your powershell
$ curl https://raw.githubusercontent.com/tensorflow/magenta/master/magenta/tools/magenta-install.sh > /tmp/magenta-install.sh
$ bash /tmp/magenta-install.sh
```

**Step 4 [ONLY IF AUTO INSTALL FAILED]:** Activate the installation.


```
$ source ~/.bashrc
```

**Step 5:** Create new Python 3.7 environment in anaconda.
~~~
$ conda create -n magenta python=3.7
~~~

**Step 6:** Switch to your newly created 'magenta' environment.
~~~
$ conda activate magenta
~~~

**Step 7:** Ensure packages are up to date before continuing.
~~~
$ sudo apt update
$ sudo apt upgrade
~~~

**Step: 8** Install magenta music dependencies.
~~~
$ sudo apt-get install build-essential libasound2-dev libjack-dev portaudio19-dev
$ pip install --pre python-rtmidi
~~~

**Step: 9** Install magenta.
~~~
#will not return 'Magenta Install Success', is successful if there are no errors
$ pip install jupyter magenta
~~~
